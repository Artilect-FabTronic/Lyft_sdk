```python
DATA_PATH = 'input/'
train = pd.read_csv(DATA_PATH + 'train.csv')
#sample_submission = pd.read_csv(DATA_PATH + 'sample_submission.csv')
lyft_dataset = LyftDataset(data_path=DATA_PATH, json_path=DATA_PATH+'train_data', map_resolution = 0.1)

```

```python
lyft_dataset.table_names
```

```python
my_scene = lyft_dataset.scene[0]
N_token = my_scene['first_sample_token']
Last_token=my_scene['last_sample_token']
Last_token
```

```python
N=0
M=0
Liste_samples_token=[]
while N<100:
    for list_sample in lyft_dataset.sample[:]:
        if N_token == Last_token:
            break

        if N_token in list_sample['token']:
            #if list_sample_data['channel']=='CAM_FRONT':
            N_token=list_sample['next']
            Liste_samples_token.append(N_token)
    N=N+1
```

```python
Liste_samples_token
```

```python
fichiers_JPEG=[]
for x in Liste_samples_token:
    for list_sample_data in lyft_dataset.sample_data[:]:
           if x in list_sample_data['sample_token']:
                if list_sample_data['channel'] == 'CAM_FRONT':
                    print(list_sample_data['filename'])
                    fichiers_JPEG.append(list_sample_data['filename'])
                
```

```python
DATA_PHOTO = "Scene_1_CAM_FRONT/"
parent_dir= 'input/'
directory= DATA_PHOTO +"images/"
path = os.path.join(parent_dir,directory) 
os.makedirs(path)
```
